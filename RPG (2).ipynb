```python
import time
import random as rd

#criação de personagem

personagem = {
    'HP' : 1500,
    'Nome': '',
    'LVL': 0,
    'Vigor': 0,
    'Mente': 0,
    'Fortitude': 0,
    'Força': 0,
    'Destreza' : 0,
    'Inteligência': 0,
    'Fé': 0,
    'Arcano': 0,
    'Runas' : 0,
    'Inventario': []
    
}
print('-'*10,'CRIAÇÃO DE PERSONAGEM', '-'*10)
personagem['Nome'] = input('Nome: ')

#escolha de classe

print('---------- ESCOLHA DE CLASSE ----------')
print('''
(A). Vagabundo: Guerreiro robusto com espada, alabarda e escudo; ótimo para iniciantes que querem resistência e força.

(B). Guerreiro: Especialista em destreza, começa com duas espadas curvas; rápido e ágil, mas pouco defensivo.

(C). Astrólogo: Mago puro, com alto intelecto e cajado; poderoso em feitiços, mas frágil fisicamente.
''')
print()
classe = input('Qual classe você deseja usar? ')

if classe == 'A' or classe == 'a':
    personagem['LVL'] = 9
    personagem['Vigor'] = 15
    personagem['Mente'] = 10
    personagem['Fortitude'] = 11
    personagem['Força'] = 14
    personagem['Destreza'] = 13
    personagem['Inteligência'] = 9
    personagem['Fé'] = 9
    personagem['Arcano'] = 7

if classe == 'B' or classe == 'b':
    personagem['LVL'] = 8
    personagem['Vigor'] = 11
    personagem['Mente'] = 12
    personagem['Fortitude'] = 11
    personagem['Força'] = 10
    personagem['Destreza'] = 16
    personagem['Inteligência'] = 10
    personagem['Fé'] = 8
    personagem['Arcano'] = 9

if classe == 'C' or classe == 'c':
    personagem['LVL'] = 6
    personagem['Vigor'] = 9
    personagem['Mente'] = 15
    personagem['Fortitude'] = 9
    personagem['Força'] = 8
    personagem['Destreza'] = 12
    personagem['Inteligência'] = 16
    personagem['Fé'] = 7
    personagem['Arcano'] = 9

def status():
    print()
    print('-----Status-----')
    for k,v in personagem.items():
        print(f'{k}:{v}')
    time.sleep(5)

listaItens = ['Blasphemous Blade', 'Malos Sword', 'Winged Scythe', 'Rivers of Blood']

def item_encontrado():
    item = rd.choice(listaItens)
    print(item)
    add = input(f'Deseja adicionar {item} ao inventário? (S/N)')
    if add == 'S' or add == 's':
       personagem['Inventario'].append(item)

def ver_inventario():
    print()
    print('-----Inventário-----')
    print()
    for item in personagem['Inventario']:
        print(f'- {item}')
    if not personagem['Inventario']:
        print()
        print('O inventário está vazio')
    print()
    time.sleep(5)

def add_item():
    print()
    nomeItem = input('Insira o nome do item: ')
    personagem['Inventario'].append(nomeItem)
    print()
    print('Item adicionado com sucesso')
    time.sleep(3)


#menu inicial

print()
print()
while True:
    print()
    print('----------MENU DE PAUSA----------')
    print()
    print('1 - Ver status')
    print('2 - Adicionar item')
    print('3 - Ver inventário')
    print('4 - Voltar ao jogo')
    print()

    opcao = int(input('Escolha uma opção: '))
    if opcao == 1:
        status()
    elif opcao == 2:
        add_item()
    elif opcao == 3:
        ver_inventario()
    elif opcao == 4:
        print()
        print('Voltando ao jogo...')
        print()
        break
    else:
        print()
        print('Opção inválida, tente novamente.')
        print()
        print('--------------------------------')


#criação de inimigo

inimigo = {
    'Nome':'Hollow',
    'HP': 1500,
    'Ataque': 100,
    'Runas' : 50
}

#sistema de batalha

def lutar():
    print()
    time.sleep(2)
    print(f'O inimigo {inimigo['Nome']} apareceu!')
    print('Entrando em batalha...')
    print()
    time.sleep(2)
    while personagem['HP'] > 0 and inimigo['HP'] > 0:
        dano = rd.randint(100,200)
        inimigo['HP'] -= dano
        personagem['HP'] -= inimigo['Ataque']
        print()
        print(f'Turno de {personagem['Nome']}: {dano} de dano!')
        print()
        time.sleep(1)
        print(f'Turno de {inimigo['Nome']}: {inimigo['Ataque']} de dano!')
        print()
    
    if personagem['HP'] <= 0:
        print()
        print('Você morreu')
        print()
    elif inimigo['HP'] <= 0:
        print()
        print(f'{inimigo['Nome']} derrotado')
        print(f'+{inimigo['Runas']} runas')
        personagem['Runas'] += inimigo['Runas']
        print()

lutar()

#subir de nivel

if personagem['Runas'] >= 100:
    print('Subiu de nível')
    print(f'{personagem['LVL']} --> {personagem['LVL'] + 1}')
    personagem['LVL'] = personagem['LVL'] + 1
    if classe == 'A' or classe == 'a':
        personagem['HP'] += personagem['HP'] * 0.10
        personagem['Força'] += personagem['Força'] * 0.10
        personagem['Fortitude'] += personagem['Fortitude'] * 0.10

    elif classe == 'B' or classe == 'b':
        personagem['HP'] += personagem['HP'] * 0.10
        personagem['Destreza'] += personagem['Destreza'] * 0.10
        personagem['Fortitude'] += personagem['Fortitude'] * 0.10

    else:
        personagem['HP'] += personagem['HP'] * 0.10
        personagem['Mente'] += personagem['Mente'] * 0.10
        personagem['Inteligência'] += personagem['Inteligência'] * 0.10
        
        
    
        
        
    
```
